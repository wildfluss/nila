```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set_style ("darkgrid")
mean_plot = 3
num = 40
```

```python
x1 = np.random.normal (mean_plot, 1, size = num)
y1 = np.random.normal (mean_plot, 1, size = num)
z1 = np.random.normal (mean_plot, 1, size = num)
```

```python
plt.figure(figsize = (5, 4))
plot_axes = plt.axes(projection = '3d')
```

```python
print(type(plot_axes))
plot_axes.scatter3D(x1, y1, z1)
```

```python
plot_axes.set_xlabel('x')
plot_axes.set_ylabel('y')
plot_axes.set_zlabel('z')
plt.show()
```

```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
def f(x):
    # return x*2
    param = x # st.session_state['param']
    def FUNC_Z(x_plot, y_plot):
        return 50 - (x_plot**param + y_plot**param)
    num = 40
    x_plot1 = np.linspace(-4, 4, num)
    y_plot1 = np.linspace(-4, 4, num)
    plot1, plot2 = np.meshgrid(x_plot1, y_plot1)
    plot3 = FUNC_Z(plot1, plot2)
    fig = plt.figure(figsize=(10, 4))
    axes = fig.add_axes([-4, -4, 4, 4], projection='3d')
    axes.plot_surface(plot1, plot2, plot3)
    plt.show ()    
```
